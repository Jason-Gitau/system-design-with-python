```python
from IPython.display import Image, display  # ✅ moved to top

from diagrams import Diagram, Cluster
from diagrams.onprem.client import User
from diagrams.onprem.compute import Server
from diagrams.onprem.network import Nginx
from diagrams.onprem.database import PostgreSQL

from diagrams.aws.compute import ECS
from diagrams.aws.integration import SQS

with Diagram(
    "MPesa System Architecture (USSD + API Flow)",
    show=False,
    filename="mpesa_architecture"  # ✅ explicit, predictable filename
):
    # Users
    mobile_user = User("Mobile User (*334#)")
    app_user = User("App / Website User")

    # --- USSD FLOW ---
    with Cluster("USSD Channel"):
        ussd_gateway = Server("USSD Gateway")
        telecom_core = Server("Safaricom Core Network")

    # --- API FLOW ---
    with Cluster("Daraja API Gateway"):
        api_gateway = Nginx("API Gateway")
        auth_service = Server("OAuth Service")

    # --- MPesa Core ---
    with Cluster("MPesa Core System"):
        payment_processor = ECS("Payment Processor")
        transaction_queue = SQS("Transaction Queue")
        worker = ECS("Processing Worker")
        mpesa_db = PostgreSQL("MPesa DB")

    # --- External Systems ---
    with Cluster("External Integrations"):
        business_server = Server("Business Backend (Your Bot/API)")
        callback_listener = Server("Callback URL")

    # USSD Flow
    mobile_user >> ussd_gateway >> telecom_core >> payment_processor

    # API Flow
    app_user >> api_gateway >> auth_service >> payment_processor

    # Internal Processing
    payment_processor >> transaction_queue >> worker
    worker >> mpesa_db

    # External API Integration (Daraja)
    payment_processor >> business_server
    worker >> callback_listener

    # Persistence
    payment_processor >> mpesa_db

# ✅ outside the with block
display(Image("mpesa_architecture.png"))
```
